```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime, timedelta

# Define the start and end dates for the project timeline
start_date = datetime(2024, 12, 1)
end_date = datetime(2025, 3, 1)

# Data for the tasks and durations
tasks = [
    ('Design wireframes', 6, 'US-1.0', start_date),
    ('Develop interface', 12, 'US-1.0', start_date + timedelta(days=6)),
    ('Usability testing', 4, 'US-1.0', start_date + timedelta(days=18)),

    ('Requirements', 6, 'US-2.0', start_date + timedelta(days=6)),
    ('Develop logic', 12, 'US-2.0', start_date + timedelta(days=12)),
    ('Test accuracy', 6, 'US-2.0', start_date + timedelta(days=24)),

    ('Analyze bugs', 4, 'US-3.0', start_date + timedelta(days=18)),
    ('Develop fixes', 4, 'US-3.0', start_date + timedelta(days=22)),

    ('Gather requirements', 8, 'US-4.0', start_date + timedelta(days=24)),
    ('Write documentation', 12, 'US-4.0', start_date + timedelta(days=32)),

    ('Set up channel', 3, 'US-5.0', start_date + timedelta(days=34)),
    ('Define workflow', 4, 'US-5.0', start_date + timedelta(days=37)),
    ('Train team', 2, 'US-5.0', start_date + timedelta(days=41)),

    ('Log structure', 4, 'US-6.0', start_date + timedelta(days=38)),
    ('Populate requests', 6, 'US-6.0', start_date + timedelta(days=42)),

    ('Map interfaces', 10, 'US-7.0', start_date + timedelta(days=40)),
    ('Develop APIs', 20, 'US-7.0', start_date + timedelta(days=50)),
    ('Test integrations', 15, 'US-7.0', start_date + timedelta(days=70)),
    ('Resolve errors', 10, 'US-7.0', start_date + timedelta(days=85)),

    ('Data collection', 4, 'US-8.0', start_date + timedelta(days=60)),
    ('Write report', 6, 'US-8.0', start_date + timedelta(days=64)),

    ('Identify features', 8, 'US-9.0', start_date + timedelta(days=50)),
    ('Develop demo', 16, 'US-9.0', start_date + timedelta(days=58)),
    ('Create presentation', 4, 'US-9.0', start_date + timedelta(days=74)),
    ('Feedback collection', 6, 'US-9.0', start_date + timedelta(days=78)),

    ('Define requirements', 8, 'US-10.0', start_date + timedelta(days=70)),
    ('Develop functionality', 20, 'US-10.0', start_date + timedelta(days=78)),
    ('Test logs', 10, 'US-10.0', start_date + timedelta(days=98)),
    ('Ensure compliance', 6, 'US-10.0', start_date + timedelta(days=108)),
]

# Create a DataFrame to hold the task data
df = pd.DataFrame(tasks, columns=['Task', 'Duration (days)', 'Story ID', 'Start Date'])

# Calculate end dates for tasks
df['End Date'] = df['Start Date'] + pd.to_timedelta(df['Duration (days)'], 'D')

# Plotting the Gantt chart
fig, ax = plt.subplots(figsize=(12, 8))

# Format date ticks
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))
plt.xticks(rotation=45)

# Plot each task as a horizontal bar
for index, row in df.iterrows():
    ax.barh(row['Task'], (row['End Date'] - row['Start Date']).days, left=row['Start Date'], color='skyblue', edgecolor='black')

# Add task labels with story ID
for index, row in df.iterrows():
    ax.text(row['Start Date'] + (row['End Date'] - row['Start Date']) / 2,
            index,
            row['Story ID'],
            va='center',
            ha='center',
            color='black',
            fontweight='bold')

# Set labels and title
ax.set_xlabel('Timeline')
ax.set_ylabel('Tasks')
ax.set_title('Gantt Chart for UltraFund Project (1st December 2024 to 1st March 2025)')

# Show gridlines and plot
ax.grid(True, axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
